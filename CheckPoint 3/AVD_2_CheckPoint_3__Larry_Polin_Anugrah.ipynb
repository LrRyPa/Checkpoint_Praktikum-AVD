## Produk atau Kategori Terlaris

### Produk Terlaris

Top 5 Produk Clothing

```python
clothing = data[data['category'] == 'Clothing']
ranking_sales = clothing[(clothing['ranking_sales'] >= 1) & (clothing['ranking_sales'] <= 50)]
clothing_sales = ranking_sales.sort_values(by = 'ranking_sales', ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(clothing_sales['product_id'].astype(str), clothing_sales['total_sales'], color='skyblue')
plt.title('5 Produk Teratas Pada Kategori Clothing')
plt.xlabel('Product ID')
plt.ylabel('total sales')
plt.xticks(rotation=45)
plt.show()
```

A. Insight:
* Penjualan kelima produk teratas memiliki jumlah total penjualan yang mirip dengan rentang antara 7.500 hingga 8.500 unit, yang menunjukkan tidak ada dominasi pada produk tertentu.

B. Action:
* Menganalisis apa yang membuat produk dengan ID 286 lebih unggul dari produk lainnya.
* Menggunakan hasil analisisnya untuk meningkatkan produk lainnya di kategori ini.

## Analisis Tren Penjualan Berdasarkan Waktu

### Tren Penjualan

```python
month = data.groupby('peak_sales_month')['category'].count().reindex([
    'sales_month_1', 'sales_month_2', 'sales_month_3', 'sales_month_4',
    'sales_month_5', 'sales_month_6', 'sales_month_7', 'sales_month_8',
    'sales_month_9', 'sales_month_10', 'sales_month_11', 'sales_month_12'
])

plt.figure(figsize=(10, 6))
plt.plot(range(1, 13), month.values, marker='o', color='blue')
plt.title('Distribusi Bulan Penjualan Puncak')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Produk')
plt.grid(alpha=0.3)
plt.xticks(range(1, 13), ['januar', 'februari', 'maret', 'april', 'mei', 'juni', 'juli', 'agustus', 'september', 'oktober', 'november', 'desember'], rotation=45)
plt.tight_layout()
plt.show()
```

A. Insight:

* Puncak penjualan tertinggi terjadi di bulan **Juli** dengan jumlah produk yang terjual lebih dari 100. Peningkatan ini  mungkin disebabkan oleh faktor musiman seperti liburan atau promosi besar besaran.
* Penjualan terendah terjadi di bulan **Desember** dengan jumlah produk sekitar 70 yang terjual. Kemungkinan ini disebabkan oleh minat konsumen yang berkurang setelah belanja besar besaran di bulan sebelumnya.

B. Action:

* Mengoptimalkan strategi pada bulan Juli dengan menambahkan stok dan pemasaran untuk memanfaatkan faktor musiman di bulan Juli.
* Mengevaluasi faktor faktor yang memengaruhi penurunan penjualan dan cara untuk meningkatkan minat pelanggan pada bulan desember seperti bundling produk.

A. Clothing dan Electronics
*  Insight:

  Dalam kategori Clothing dan Electronics ada beberapa outlier dengan jumlah ulasan yang jauh lebih tinggi daripada median, yang menunjukkan bahwa produk produk tertentu dalam kategori tersebut sangat diminati.
*  Action:

  Menganalisis  karakteristik produk pada kedua kategori tersebut yang memiliki ulasan tinggi yang kemudian dapat digunakan untuk membuat produk lainnya sukses.

B. Home & Kitchen, Health dan Sports
* Insight:

  Distribusi ulasan ketiga kategori ini memiliki rentang interkuartil yang lebih luas, yang menunjukkan variasi besar dalam jumlah ulasan yang diterima produk dalam kategori tersebut.

* Action:

  Meningkatkan kembali kualitas agar tetap konsisten dan terus mengelola ekspektasi pelanggan agar mempertahankan distribusi ulasan tetap stabil.

C. Books dan Toys
* Insight:

  Median dan pesebaran ulasan pada kategori ini yang lebih baik ketimbang kategori lainnya. ini menunjukkan bahwa pelanggan cukup puas dengan kedua kategori ini,meskipun pada kategori toys terdapat beberapa outlier.

* Action:

  Meningkatkan pemasaran produk yang memiliki ulasan tinggi serta menggunakan ulasan positif sebagai daya tarik pelanggan.
